\1. Breast Cancer Wisconsin (Diagnostic)— Data Preprocessing

```python
df = pd.read_csv('/content/drive/MyDrive/dataset/data.csv')

print("Shape:", df.shape)
df.info()
df.describe()
```

Data Loading and Initial Inspection:The dataset contains 569 observations and 33 columns. The target diagnosis is a text label; the remaining variables are numeric measurements of cell nuclei.

```python
empty_cols = df.columns[df.isnull().all()].tolist()
df.drop(columns=[c for c in ['id'] + empty_cols if c in df.columns], inplace=True)

missing = df.isnull().sum()
print("Missing values:\n", missing[missing > 0] if missing.any() else "None")

print("Duplicate rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)

print("Shape after cleaning:", df.shape)
```

Data Cleaning: `id`  and `Unnamed: 32`  were dropped, taking the dataset from (569, 33) to (569, 31). No missing values or duplicate rows remain , the dataset is clean and ready for transformation.

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
print("Mapping:", dict(zip(le.classes_, le.transform(le.classes_))))
print(df['diagnosis'].value_counts())

plt.figure(figsize=(5, 4))
sns.countplot(x='diagnosis', data=df)
plt.xticks([0, 1], ['Benign (0)', 'Malignant (1)'])
plt.title("Diagnosis Distribution")
plt.xlabel("Diagnosis")
plt.ylabel("Number of Patients")
plt.show()
```

Encoding the Target Variable:`diagnosis` was label-encoded (B → 0, M → 1), giving 357 Benign cases (62.7%) and 212 Malignant cases (37.3%) — a mild, manageable class imbalance that does not require synthetic oversampling.

```python
bin_features = ['radius_mean', 'texture_mean', 'area_mean']

for feat in bin_features:
    df[f'{feat}_category'] = pd.cut(df[feat], bins=3, labels=['Small', 'Medium', 'Large'])

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, feat in zip(axes, bin_features):
    sns.countplot(x=f'{feat}_category', data=df, order=['Small', 'Medium', 'Large'], ax=ax)
    ax.set_title(feat)
    ax.set_xlabel('Category')
    ax.set_ylabel('Frequency')
plt.tight_layout()
plt.show()

for feat in bin_features:
    print(df[f'{feat}_category'].value_counts())
```

Binning numerical features:`radius_mean`, `texture_mean`, and `area_mean` were split into Small/Medium/Large bins. All three are concentrated in the Small to Medium range with a sharp drop-off at Large, showing that extreme cellular overgrowth is uncommon . A useful, intuitive view for diagnostic risk stratification.

```python
outlier_cols = ['area_mean', 'radius_mean', 'perimeter_mean', 'texture_mean', 'concavity_mean']
bounds = {}

for col in outlier_cols:
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    bounds[col] = (lower, upper)
    n_out = ((df[col] < lower) | (df[col] > upper)).sum()
    print(f"{col}: {n_out} outliers  (bounds {lower:.2f} to {upper:.2f})")

fig, axes = plt.subplots(1, len(outlier_cols), figsize=(18, 3))
for ax, col in zip(axes, outlier_cols):
    sns.boxplot(x=df[col], ax=ax)
    ax.set_title(f'Before: {col}')
plt.tight_layout()
plt.show()

# Cap using the bounds already computed above (no recalculation needed)
for col, (lower, upper) in bounds.items():
    df[col] = df[col].clip(lower, upper)

fig, axes = plt.subplots(1, len(outlier_cols), figsize=(18, 3))
for ax, col in zip(axes, outlier_cols):
    sns.boxplot(x=df[col], ax=ax)
    ax.set_title(f'After: {col}')
plt.tight_layout()
plt.show()
```

Outlier Detection and Capping:Using the IQR method, `area_mean` with 25 outliers, followed by `concavity_mean`  with 18, `radius_mean` with 14, `perimeter_mean`  with 13, and `texture_mean` with 7. All extreme values were capped to their IQR bounds rather than removed, so all 569 observations are preserved while the influence of extreme tumor sizes and shapes is reduced. The "after" boxplots confirm the extreme values are now compact within the whiskers.

```python
dist_features = ['radius_mean', 'texture_mean', 'area_mean', 'concavity_mean']

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for ax, feat in zip(axes.flat, dist_features):
    sns.histplot(df[feat], kde=True, bins=20, ax=ax)
    ax.set_title(f'Distribution of {feat}')
plt.tight_layout()
plt.show()

df[dist_features].describe()
```

 Distribution Analysis (Histogram + KDE):`radius_mean`, `area_mean`, and `concavity_mean` are all positively (right) skewed — most tumors are small with a long tail of larger, more irregular ones. `texture_mean` is closer to symmetric with the least variability. `area_mean` shows the strongest skew (skewness ≈ 1.65), reinforcing that a small number of unusually large tumors drive most of the variance and the same features flagged as outlier-heavy above. These distribution shapes are consistent with what distinguishes benign from malignant cases and are informative for later classification modeling.

2. Iris Dataset — Data Preprocessing

```python
df = pd.read_csv('/content/drive/MyDrive/dataset/Iris.csv')

print("Shape:", df.shape)
df.info()
df.describe()
```

Data Loading and Initial Inspection

```python
df.drop(columns=[c for c in ['Id'] if c in df.columns], inplace=True)

print("Missing values:\n", df.isnull().sum())
print("\nDuplicate rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)

print("\nShape after cleaning:", df.shape)
df.head()
```

Data Cleaning:After dropping the `Id` column and removing duplicate rows, the dataset has no missing values and is ready for transformation.

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Species'] = le.fit_transform(df['Species'])
print("Mapping:", dict(zip(le.classes_, le.transform(le.classes_))))
df['Species'].value_counts()
```

Encoding `Species`:`Species` was converted from text into numeric labels, with each of the three species assigned a unique integer, making the dataset compatible with machine learning algorithms that require numeric input.

```python
df['SepalLength_Category'] = pd.cut(
    df['SepalLengthCm'], bins=3, labels=['Small', 'Medium', 'Large']
)
df['SepalLength_Category'].value_counts()
```

Binning :`SepalLengthCm` was grouped into Small/Medium/Large bins, turning the continuous measurement into an easily interpretable size category.

```python
numeric_features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
for ax, col in zip(axes.flat, numeric_features):
    sns.boxplot(x=df[col], ax=ax)
    ax.set_title(f'Before: {col}')
plt.tight_layout()
plt.show()

for col in numeric_features:
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    n_out = ((df[col] < lower) | (df[col] > upper)).sum()
    print(f"{col}: {n_out} outliers")
    df[col] = df[col].clip(lower, upper)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
for ax, col in zip(axes.flat, numeric_features):
    sns.boxplot(x=df[col], ax=ax)
    ax.set_title(f'After: {col}')
plt.tight_layout()
plt.show()
```

Outlier Detection and Capping:Using the IQR method, 0 outliers were found in `SepalLengthCm`, 4 in `SepalWidthCm`, and 0 in both `PetalLengthCm` and `PetalWidthCm`. The 4 extreme values in `SepalWidthCm` were capped to the IQR bounds rather than removed, preserving all 147 observations while limiting the influence of extreme values.

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for ax, col in zip(axes.flat, numeric_features):
    sns.histplot(df[col], kde=True, bins=20, ax=ax)
    ax.set_title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

Distribution Analysis (Histogram + KDE):`SepalLengthCm` and `SepalWidthCm` are roughly bell-shaped with slight skew, showing most observations clustered near the mean. `PetalLengthCm` and `PetalWidthCm` are multimodal, reflecting the three distinct Iris species — this indicates petal measurements separate the species more clearly than sepal measurements, making them more informative for classification.
